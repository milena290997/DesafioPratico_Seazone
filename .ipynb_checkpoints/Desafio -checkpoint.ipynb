Lendo as Bibliotecas necessárias para o teste

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

Vamos ler o arquivo que sera usado no teste

```python
desafio = pd.read_csv('desafio_ratings.csv',encoding='latin-1',delimiter=';')
desafio.head(5)
```

```python
desafio_detalhes = pd.read_csv('desafio_details.csv',encoding='latin-1',delimiter=';')
desafio_detalhes.head(5)
```

# 1. Ordene as cidades em ordem crescente de número de listings;


```python
# Contar o número de listings por cidade
city_listings_count = desafio_detalhes['city_name'].value_counts().reset_index()
city_listings_count.columns = ['city_name', 'listings_count']

# Ordenar as cidades em ordem crescente de número de listings
sorted_cities = city_listings_count.sort_values('listings_count')

print(sorted_cities)
```

# 2. Ordene as cidades em ordem decrescente de metros quadrados

```python
# Ordenar o DataFrame em ordem decrescente pela coluna 'room_surface_in_m2'
df_ordenado = desafio_detalhes.sort_values(by='room_surface_in_m2', ascending=False)

# Mostrar as cidades em ordem decrescente de metros quadrados
cidades_ordem_decrescente = df_ordenado['city_name'].unique()
print(cidades_ordem_decrescente)
```

# 3. Quais cidades têm mais avaliações?

```python
# Agrupar o DataFrame de avaliações por hotel_id e somar o número total de avaliações
rating_grouped = desafio.groupby('hotel_id')['number_of_ratings'].sum().reset_index()

# Combinar os DataFrames de detalhes e avaliações usando hotel_id
combined_df = pd.merge(desafio_detalhes, rating_grouped, on='hotel_id', how='left')

# Agrupar por city_name e calcular o número total de avaliações por cidade
city_ratings_count = combined_df.groupby('city_name')['number_of_ratings'].sum().reset_index()

# Ordenar as cidades em ordem decrescente de número total de avaliações
sorted_cities_by_ratings = city_ratings_count.sort_values('number_of_ratings', ascending=False)

print(sorted_cities_by_ratings)
```

# 4. Quais cidades têm a maior média de avaliações? E a menor média?
# a. Existe alguma explicação para isso? Conjecture.


#### Quanto a explicação para isso: Qualidade das instalações: Cidades com acomodações de alta qualidade e comodidades superiores podem ter médias de avaliação mais altas; Atendimento ao cliente: A satisfação dos hóspedes também pode depender da qualidade do serviço prestado; Localização: Cidades com acomodações bem localizadas em relação a atrações turísticas, transporte e conveniências podem receber avaliações melhores; Variedade de acomodações: Cidades com uma variedade maior de tipos de acomodações podem ter uma distribuição mais ampla de avaliações.

```python
# Agrupar o DataFrame de avaliações por hotel_id e calcular a média do número de avaliações
rating_grouped = desafio.groupby('hotel_id')['number_of_ratings'].sum().reset_index()

# Combinar os DataFrames de detalhes e avaliações usando hotel_id
combined_df = pd.merge(desafio_detalhes, rating_grouped, on='hotel_id', how='left')

# Agrupar por city_name e calcular a média do número de avaliações por cidade
city_avg_ratings = combined_df.groupby('city_name')['number_of_ratings'].mean().reset_index()

# Ordenar as cidades em ordem decrescente de média de avaliações
sorted_cities_by_avg_ratings = city_avg_ratings.sort_values('number_of_ratings', ascending=False)

# Identificar as cidades com maior e menor média de avaliações
cidade_maior_media = sorted_cities_by_avg_ratings.iloc[0]
cidade_menor_media = sorted_cities_by_avg_ratings.iloc[-1]

print("Cidade com maior média de avaliações:", cidade_maior_media['city_name'])
print("Média de avaliações:", cidade_maior_media['number_of_ratings'])

print("\nCidade com menor média de avaliações:", cidade_menor_media['city_name'])
print("Média de avaliações:", cidade_menor_media['number_of_ratings'])
```

# 5. Existem correlações entre as características de um anúncio e a sua localização?
# a. Quais? Explique

#### Correlação possivel Tamanho do quarto (room_surface_in_m2) e Localização (city_name) 

```python
# Unir os DataFrames 'ratings_data' e 'details_data' usando 'hotel_id' como chave
merged_data = pd.merge(desafio, desafio_detalhes, on='hotel_id')

# Calcular a correlação entre 'city_name' (localização) e 'room_surface_in_m2' (tamanho do quarto)
correlation_surface_location = merged_data.groupby('city_name')['room_surface_in_m2'].mean().reset_index()

# Exibir a média do tamanho do quarto por localização
print(correlation_surface_location)

```

#  6. Existem relações entre a nota do anúncio e os recursos disponíveis no imóvel?
 # a. Quais? Explique

#### A explicação para as relações encontradas pode variar:

#### Correlação positiva: Se houver uma correlação positiva entre a presença de certos recursos e as notas de avaliação, isso pode indicar que hóspedes valorizam esses recursos e tendem a dar notas mais altas quando estão disponíveis.

#### Correlação negativa: Já uma correlação negativa poderia indicar que a presença de alguns recursos pode impactar negativamente a avaliação, possivelmente devido a problemas com esses recursos.

#### Sem correlação: A falta de correlação pode sugerir que a presença ou ausência de determinados recursos pode não ter influência direta nas avaliações dos hóspedes.

```python
# Combinar os DataFrames usando hotel_id como chave primária
merged_df = pd.merge(desafio, desafio_detalhes, on='hotel_id', how='inner')

# Analisar os recursos disponíveis nos imóveis
merged_df['room_facilities'] = merged_df['room_facilities'].str.split(',')

# Calcular a correlação entre notas de avaliação e recursos disponíveis
# Por exemplo, considerando a correlação de Pearson
correlation_matrix = merged_df[['Comodidades', 'Limpeza', 'Conforto', 'Custo-benefício', 'Localização', 'Total', 'WiFi gratuito']].corr()

# Visualizar a matriz de correlação em um mapa de calor
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlação entre Notas de Avaliação')
plt.show()

# Exemplo de gráfico de dispersão entre Comodidades e Nota Total
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Comodidades', y='Total', data=merged_df)
plt.title('Relação entre Comodidades e Nota Total')
plt.xlabel('Comodidades')
plt.ylabel('Nota Total')
plt.show()

# Exemplo de boxplot para avaliar a distribuição das notas com base na presença/ausência de WiFi gratuito
plt.figure(figsize=(8, 6))
sns.boxplot(x='WiFi gratuito', y='Total', data=merged_df)
plt.title('Distribuição da Nota Total com WiFi gratuito')
plt.xlabel('WiFi gratuito')
plt.ylabel('Nota Total')
plt.show()
```

#  7. Existe alguma relação entre a nota recebida e a localização?
 # a. Quais? Explique

#### Correlação positiva: Se houver uma correlação positiva entre a localização e a nota de avaliação, isso pode sugerir que os hóspedes tendem a dar notas mais altas para acomodações localizadas em áreas que oferecem acesso conveniente a atrações, transporte público, restaurantes, etc.

#### Correlação negativa: Uma correlação negativa pode indicar que a localização pode ser um ponto problemático para os hóspedes, afetando negativamente suas avaliações. Isso pode incluir áreas barulhentas, distantes de pontos turísticos ou com problemas de segurança.

#### Sem correlação aparente: A falta de correlação pode sugerir que, para os hóspedes, a localização não é um fator significativo ao avaliar acomodações.

```python
# Calculando a correlação entre a nota total e a pontuação de localização
correlation_total_location = merged_df['Total'].corr(merged_df['Localização'])

# Exibindo a correlação
print(f"Correlação entre Nota Total e Localização: {correlation_total_location}")

# Gráfico de dispersão entre Nota Total e Localização
plt.figure(figsize=(8, 6))
sns.scatterplot(x='Localização', y='Total', data=merged_df)
plt.title('Relação entre Nota Total e Localização')
plt.xlabel('Localização')
plt.ylabel('Nota Total')
plt.show()

# Boxplot para comparar a distribuição das notas por localização
plt.figure(figsize=(10, 6))
sns.boxplot(x='city_name', y='Total', data=merged_df)
plt.title('Distribuição da Nota Total por Localização')
plt.xlabel('Localização (Cidade)')
plt.ylabel('Nota Total')
plt.xticks(rotation=45)
plt.show()
```

# 8. O que você pode inferir sobre as notas dos imóveis?


```python
# Estatísticas descritivas das notas
desc_stats = merged_df[['Comodidades', 'Limpeza', 'Conforto', 'Custo-benefício', 'Localização', 'WiFi gratuito', 'Total']].describe()
print(desc_stats)

# Distribuição das notas (exemplo para a nota total)
plt.figure(figsize=(8, 6))
sns.histplot(merged_df['Total'], bins=20, kde=True)
plt.title('Distribuição da Nota Total')
plt.xlabel('Nota Total')
plt.ylabel('Frequência')
plt.show()

# Comparação das notas por categorias
plt.figure(figsize=(10, 6))
sns.boxplot(data=merged_df[['Comodidades', 'Limpeza', 'Conforto', 'Custo-benefício', 'Localização', 'WiFi gratuito']])
plt.title('Comparação das Notas por Categoria')
plt.xlabel('Categoria')
plt.ylabel('Nota')
plt.xticks(rotation=45)
plt.show()

# Correlação entre o número de avaliações e as notas
correlation_ratings_total = merged_df['number_of_ratings'].corr(merged_df['Total'])
print(f"Correlação entre Número de Avaliações e Nota Total: {correlation_ratings_total}")
```

# 9. Quais são os anúncios que te parecem críticos?
 # a. Explique.

#### Olhando para as médias, os valores que se destacam como críticos são os que estão substancialmente abaixo da média. Por exemplo, 6136321 (3.0), 7597212 (3.3), 7620153 (1.0), entre outros com notas consideravelmente inferiores às médias correspondentes de suas categorias.

#### Esses anúncios podem ser considerados críticos, pois têm avaliações muito baixas em comparação com as médias estabelecidas para cada categoria, o que pode indicar áreas problemáticas nos serviços ou instalações oferecidos pelos hotéis

```python
# Média das notas por hotel_id (anúncio)
mean_ratings = merged_df.groupby('hotel_id')['Total'].mean()

# Identificar anúncios com notas consideravelmente baixas (exemplo: notas menores que 6)
critical_ads = mean_ratings[mean_ratings < 6]

# Visualizar os anúncios considerados críticos
print("Anúncios críticos:")
print(critical_ads)

# Exemplo de identificação de recursos com baixas avaliações recorrentes
# Aqui, vou verificar a média das notas para cada recurso
mean_scores_by_feature = merged_df[['Comodidades', 'Limpeza', 'Conforto', 'Custo-benefício', 'Localização', 'WiFi gratuito']].mean()

# Visualizar médias das notas por recurso
print("\nMédia das notas por recurso:")
print(mean_scores_by_feature)
```
